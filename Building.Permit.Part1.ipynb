# Import The Library

```python
import pandas as pd
```

# Import The Dataset

```python
df = pd.read_csv('data/building_permits.csv')
df.head(3)
```

# Clean The Columns

```python
df = df.drop(columns=['PERMIT NUMBER 1', 'PERMIT NUMBER 2', 'PERMIT NUMBER 3', 'PERMIT NUMBER 4', 'PERMIT NUMBER 5'])

# drop the coloumn, 'Total'!
# df.drop(['PERMIT NUMBER 1', 'PERMIT NUMBER 2', 'PERMIT NUMBER 3', 'PERMIT NUMBER 4', 'PERMIT NUMBER 5'], )
# df.drop(['C', 'D'], axis = 1) 
```

# Check The Columns

```python
df.columns
```

# Rename The Columns

```python
df.rename(columns={"PERMIT NUMBER": "PermitNo", "PERMIT TYPE": "Type", "PERMIT SUB-TYPE": "SubType", "PERMIT STATUS": "Status", "WORK DESCRIPTION": "Desc", "PERMIT ISSUE DATE": "Date", "ESTIMATED VALUE OF WORK": "EstValue", "STATE VALUATION": "StateValue", "CONTRACTOR": "Contractor"}, inplace=True)
```
